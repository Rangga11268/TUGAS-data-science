```python
# Menentukan nilai minimal dan maksimal dari kolom 'jml_pinjaman'
min_jml_pinjaman = data['jml_pinjaman'].min()
max_jml_pinjaman = data['jml_pinjaman'].max()
print(f"Nilai minimal jml_pinjaman: {min_jml_pinjaman}")
print(f"Nilai maksimal jml_pinjaman: {max_jml_pinjaman}")

bars = plt.bar(['Min Pinjaman', 'Max Pinjaman'], [min_jml_pinjaman, max_jml_pinjaman], color=['skyblue', 'orange'])
plt.ylabel('Jumlah Pinjaman')
plt.title('Nilai Minimal dan Maksimal Jumlah Pinjaman')

plt.text(0, min_jml_pinjaman, f'{min_jml_pinjaman:,.0f}', ha='center', va='bottom')
plt.text(1, max_jml_pinjaman, f'{max_jml_pinjaman:,.0f}', ha='center', va='bottom')

plt.show()
```

```python
# Menghitung standar deviasi kolom 'jml_pinjaman' (selain 'umur')
std_jml_pinjaman = data['jml_pinjaman'].std()
print(f"Standar deviasi jml_pinjaman: {std_jml_pinjaman:.2f}")

# Visualisasi distribusi 'jml_pinjaman' dengan garis rata-rata dan std
plt.figure(figsize=(8, 5))
plt.hist(data['jml_pinjaman'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(data['jml_pinjaman'].mean(), color='orange', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(data['jml_pinjaman'].mean() + std_jml_pinjaman, color='red', linestyle='dotted', linewidth=2, label='+1 Std')
plt.axvline(data['jml_pinjaman'].mean() - std_jml_pinjaman, color='red', linestyle='dotted', linewidth=2, label='-1 Std')
plt.xlabel('Jumlah Pinjaman')
plt.ylabel('Frekuensi')
plt.title('Distribusi Jumlah Pinjaman dan Standar Deviasi')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Menghitung quantile untuk kolom 'jml_pinjaman' (selain 'umur')
quantiles = data['jml_pinjaman'].quantile([0, 0.25, 0.5, 0.75, 1.0])
print(quantiles)

# Visualisasi quantile
plt.figure(figsize=(8, 5))
plt.hist(data['jml_pinjaman'], bins=20, color='lightblue', edgecolor='black', alpha=0.7)
for q, value in quantiles.items():
    plt.axvline(value, linestyle='--', label=f'{int(q*100)}% Quantile: {value:,.0f}')
plt.xlabel('Jumlah Pinjaman')
plt.ylabel('Frekuensi')
plt.title('Distribusi dan Quantile Jumlah Pinjaman')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# data outlier dari jml_peminjaman
col_outlier = 'jml_pinjaman'

# Hitung batas bawah dan atas menggunakan IQR
Q1 = data[col_outlier].quantile(0.25)
Q3 = data[col_outlier].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Deteksi outlier
outliers = data[(data[col_outlier] < lower_bound) | (data[col_outlier] > upper_bound)]

plt.figure(figsize=(12, 4))
plt.scatter(data.index, data[col_outlier], color='skyblue', label='Data', alpha=0.7)
plt.scatter(outliers.index, outliers[col_outlier], color='red', label='Outlier', s=60, edgecolor='black')
plt.axhline(lower_bound, color='orange', linestyle='--', label='Lower Bound')
plt.axhline(upper_bound, color='orange', linestyle='--', label='Upper Bound')
plt.xlabel('Index')
plt.ylabel('Jumlah Pinjaman')
plt.title('Outlier pada Jumlah Pinjaman')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)

extreme = outliers[outliers[col_outlier] > quantiles[1.00]*0.9]
for idx, row in extreme.iterrows():
    plt.annotate(f"{int(row[col_outlier]):,}", (idx, row[col_outlier]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='darkred')

plt.tight_layout()
plt.show()
```

```python
median_saldo = data['saldo_nominatif'].median()
print(f"Median saldo_nominatif: {median_saldo:,.2f}")

plt.figure(figsize=(8, 5))
plt.hist(data['saldo_nominatif'], bins=20, color='lightblue', edgecolor='black', alpha=0.7)
plt.axvline(median_saldo, color='red', linestyle='dashed', linewidth=2, label=f'Median: {median_saldo:,.0f}')
plt.xlabel('Saldo Nominatif')
plt.ylabel('Frekuensi')
plt.title('Distribusi Saldo Nominatif')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
mode_angsuran = data['jml_angsuran_per_bulan'].mode()[0]
print(f"Modus jml_angsuran_per_bulan: {mode_angsuran}")

plt.figure(figsize=(8, 5))
plt.hist(data['jml_angsuran_per_bulan'], bins=20, color='lightblue', edgecolor='black', alpha=0.7)
plt.axvline(mode_angsuran, color='red', linestyle='dashed', linewidth=2, label=f'Modus: {mode_angsuran:,.2f}')
plt.xlabel('Jumlah Angsuran per Bulan')
plt.ylabel('Frekuensi')
plt.title('Distribusi Jumlah Angsuran per Bulan')
plt.legend()
plt.tight_layout()
plt.show()
```
