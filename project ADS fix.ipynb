```python
import numpy as np

np.random.seed(0)
data['tahun_pinjam'] = np.random.choice(range(2018, 2024), size=len(data))
data['tanggal_pinjam'] = [
    pd.Timestamp(f"{tahun}") + pd.to_timedelta(np.random.randint(0, 365), unit='D')
    for tahun in data['tahun_pinjam']
]
data.head()
```

```python
print(data.umur.unique())
print(data.umur.nunique())
```

```python
searchfor = ['-43']
data[data.umur.astype(str).str.contains('|'.join(searchfor))].head(1)
```

```python
data['umur'] = pd.to_numeric(data['umur'], errors='coerce')
data['umur'] = data['umur'].replace(-43, 43)
```

1. ringkasan statistik kolom umur

```python
min_umur = data['umur'].min()
max_umur = data['umur'].max()
print(f"Nilai minimal umur: {min_umur}")
print(f"Nilai maksimal umur: {max_umur}")

bars = plt.bar(['Min Umur', 'Max Umur'], [min_umur, max_umur], color=['skyblue', 'orange'])
plt.ylabel('Umur')
plt.title('Nilai Minimal dan Maksimal Umur Nasabah')

plt.text(0, min_umur, f'{min_umur}', ha='center', va='bottom')
plt.text(1, max_umur, f'{max_umur}', ha='center', va='bottom')

plt.show()

percentile_25 = np.percentile(data['umur'], 25)
percentile_50 = np.percentile(data['umur'], 50)
percentile_75 = np.percentile(data['umur'], 75)

plt.figure(figsize=(6,4))
plt.hist(data['umur'], bins=15, color='skyblue', edgecolor='black')
plt.axvline(percentile_25, color='orange', linestyle='dashed', linewidth=2, label=f'25th percentile: {percentile_25:.2f}')
plt.axvline(percentile_50, color='green', linestyle='dashed', linewidth=2, label=f'50th percentile: {percentile_50:.2f}')
plt.axvline(percentile_75, color='purple', linestyle='dashed', linewidth=2, label=f'75th percentile: {percentile_75:.2f}')
plt.xlabel('Umur')
plt.ylabel('Jumlah Nasabah')
plt.title('Distribusi Umur Nasabah dengan Percentile')
plt.legend()
plt.show()


rata_rata_umur = data['umur'].mean()
print(f"Rata-rata umur nasabah: {rata_rata_umur:.2f}")

plt.figure(figsize=(6,4))
plt.hist(data['umur'], bins=15, color='skyblue', edgecolor='black')
plt.axvline(rata_rata_umur, color='red', linestyle='dashed', linewidth=2, label=f'Rata-rata: {rata_rata_umur:.2f}')
plt.xlabel('Umur')
plt.ylabel('Jumlah Nasabah')
plt.title('Distribusi Umur Nasabah (mean)')
plt.legend()
plt.show()


median_umur = data['umur'].median()
print(f"Median umur nasabah: {median_umur}")

plt.figure(figsize=(6,4))
plt.hist(data['umur'], bins=15, color='skyblue', edgecolor='black')
plt.axvline(median_umur, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_umur}')
plt.xlabel('Umur')
plt.ylabel('Jumlah Nasabah')
plt.title('Distribusi Umur Nasabah (Median)')
plt.legend()
plt.show()

std_umur = data['umur'].std()
print(f"Standar Deviasi umur nasabah: {std_umur:.2f}") # type: ignore

plt.figure(figsize=(6,4))
plt.hist(data['umur'], bins=15, color='skyblue', edgecolor='black')
plt.axvline(std_umur, color='purple', linestyle='dashed', linewidth=2, label=f'Std Deviasi: {std_umur:.2f}') # type: ignore
plt.xlabel('Umur')
plt.ylabel('Jumlah Nasabah')
plt.title('Distribusi Umur Nasabah (Standar Deviasi)')
plt.legend()
plt.show()
```
