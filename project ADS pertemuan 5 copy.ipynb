```python
data = data.drop(['col'], axis = 1)
data.info()
```

```python
data['Gender'] = data['jenis_kelamin'].str.split().str.get(0)
data['Detail'] = (
    data['jenis_kelamin'].str.split().str.get(1).fillna(' ') + " " +
    data['jenis_kelamin'].str.split().str.get(2).fillna(' ')
).str.strip()
print(data[['jenis_kelamin', 'Gender', 'Detail']].head())
```

```python
print(data.umur.unique())
print(data.umur.nunique())
```

```python
searchfor = ['-43']
data[data.umur.astype(str).str.contains('|'.join(searchfor))].head(1)
```

```python
data['umur'] = pd.to_numeric(data['umur'], errors='coerce')
data['umur'] = data['umur'].replace(-43, 43)
```

```python
searchfor = ['43']
data[data.umur.astype(str).str.contains('|'.join(searchfor))].head(1)
```

```python
import matplotlib.pyplot as pit
data.plot(kind='scatter', x='umur', y='tunggakan_pokok', color='orange')
pit.xlabel('Umur')
pit.ylabel('Tunggakan Pokok')
pit.title('Scatter Plot Umur vs Tunggakan Pokok')
pit.show()
```

```python
data['umur'].plot(kind='hist', bins=15, color='lightgreen', edgecolor='black')
pit.xlabel('Umur')
pit.ylabel('Frekuensi')
pit.title('Distribusi Umur Nasabah')
pit.show()
```

```python
data['jml_pinjaman'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
pit.xlabel('Jumlah Pinjaman')
pit.ylabel('Frekuensi')
pit.title('Distribusi Jumlah Pinjaman')
pit.show()
```

```python
data['jenis_kelamin'].value_counts().plot.pie(autopct='%1.0f%%', ylabel='Jenis Kelamin')
```

```python
data['status kredit'].value_counts().plot.pie(
	autopct='%1.0f%%',
	ylabel='Status Kredit'
)
```

```python
import matplotlib.pyplot as plt

data.groupby('tahun_pinjam')['jml_pinjaman'].mean().plot(kind='line')
plt.xlabel('Tahun Pinjam')
plt.ylabel('Rata-rata Jumlah Pinjaman')
plt.title('Rata-rata Jumlah Pinjaman per Tahun Pinjam')
plt.show()
```
